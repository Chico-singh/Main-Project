```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)

print("Libraries Imported Successfully")
```

```python
expenses = pd.read_csv("Expenses_clean.csv")
income = pd.read_csv("Income_clean.csv")
```

```python
expenses.head()
income.head()
```

```python
print("Expense Dataset Shape :", expenses.shape)
print("Income Dataset Shape :", income.shape)
```

```python
print(expenses.columns)
print(income.columns)
```

```python
expenses.isnull().sum()
income.isnull().sum()
```

```python
print("Expense Duplicates :", expenses.duplicated().sum())
print("Income Duplicates :", income.duplicated().sum())
```

```python
expenses.describe()
income.describe()
```

```python
expenses["date_time"] = pd.to_datetime(expenses["date_time"])
income["date_time"] = pd.to_datetime(income["date_time"])
```

```python
expenses.dtypes
income.dtypes
```

```python
expenses["Year"] = expenses["date_time"].dt.year
expenses["Month"] = expenses["date_time"].dt.month
expenses["Month_Name"] = expenses["date_time"].dt.month_name()
expenses["Day"] = expenses["date_time"].dt.day
expenses["Day_Name"] = expenses["date_time"].dt.day_name()
expenses["Weekday"] = expenses["date_time"].dt.weekday
```

```python
income["Year"] = income["date_time"].dt.year
income["Month"] = income["date_time"].dt.month
income["Month_Name"] = income["date_time"].dt.month_name()
income["Day"] = income["date_time"].dt.day
income["Day_Name"] = income["date_time"].dt.day_name()
income["Weekday"] = income["date_time"].dt.weekday
```

```python
expenses.head()
income.head()
```

```python
expenses.to_csv("Expenses_Processed.csv", index=False)
income.to_csv("Income_Processed.csv", index=False)

print("Processed datasets saved successfully!")
```
